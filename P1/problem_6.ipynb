```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

    def __repr__(self):
        return str(self.value)


class LinkedList:
    def __init__(self):
        self.head = None

    def __str__(self):
        cur_head = self.head
        out_string = ""
        while cur_head:
            out_string += str(cur_head.value) + " -> "
            cur_head = cur_head.next
        return out_string


    def append(self, value):

        if self.head is None:
            self.head = Node(value)
            return

        node = self.head
        while node.next:
            node = node.next

        node.next = Node(value)
        
    def to_list_of_values(self):
        value_list = []
        node = self.head
        while node:
            value_list.append(node.value)
            node = node.next
        return value_list  
        

    def size(self):
        return len(self.to_list_of_values()) 

def union(llist_1, llist_2):
   
    llist_1_values = llist_1.to_list_of_values()    
    llist_2_values = llist_2.to_list_of_values()   
    union_elements = set(llist_1_values+llist_2_values)
    llist_3 = LinkedList()
    for element in union_elements:
        llist_3.append(element)
    return llist_3
    

def intersection(llist_1, llist_2):
    
    llist_1_values = llist_1.to_list_of_values()    
    llist_2_values = llist_2.to_list_of_values()   
    common_elements = set(llist_1_values)&set(llist_2_values)
    llist_3 = LinkedList()
    for element in common_elements:
        llist_3.append(element)
    return llist_3
    
```

## Time complexity

All time complexity is for worst case

### Time complexity of union() involves calculating Time complexity of 

1. to_list_of_values() of LinkedList Class which is of O(n) where n is the number of elements in a linked list1
2. to_list_of_values() of LinkedList Class which is of O(m) where m is the number of elements in a linked list2
3. list concat operation which is of O(m) where m is the number of elements in the second linkedlist
4. creating a set from the combined elements of both the linked list which is O(n+m)) where n and m respectively are the number of elements in linkedlist1,linkedlist.
5. Adding the elements of set obtained in step-4 to a new linked list which is of O(k) where k is the number of union elements and k<=n+m
 

hence total time complexity = O(n)+O(m)+O(m)+O(n+m)+O(k) = O(n+m) and to simplify becomes O(n) time operation

### Time complexity of intersection() involves calculating Time complexity of 

1. to_list_of_values() of LinkedList Class which is of O(n) where n is the number of elements in a linked list1
2. to_list_of_values() of LinkedList Class which is of O(m) where m is the number of elements in a linked list2
2. creating a set from the the elements of linkedlist1 is of O(n) where n is the number of elements in linkedlist1
3. creating a set from the the elements of linkedlist2 is of O(m) where m is the number of elements in linkedlist2
4. set intersection operation has a time complexity of O(len(set1)*len(set2))
5. adding the elements of set obtained in step-5 to a new linked list which is of O(k) where k is the common elements and in the worst case = min(len(set1),len(set2))

hence total complexity = O(n)+O(m)+O(n)+O(m)+O($len(set1)*len(set2)$)+O(k) = O(n)+O(m)+O($len(set1)*len(set2)$) after ignoring O(k) compared to other terms




## Space complexity

All space complexity is for worst case

### space complexity of union() involves calculating space complexity of 

1. memory to hold list of linkedlist1 elements = O(n) where n is the number of elements in a linked list1
2. memory to hold list of linkedlist2 elements = O(m) where m is the number of elements in a linked list2
3. memory to hold concat of list1 and list2 = O(n+m)
4. memory to hold set of concat of list1 and list2 = O(k) where k is the union elements and k <=n+m = O(n+m)

so total space complexity =  O(n)+O(m)+O(n+m)+O(k)

### space complexity of intersection() involves calculating space complexity of 

1. memory to hold list of linkedlist1 elements = O(n) where n is the number of elements in a linked list1
2. memory to hold list of linkedlist2 elements = O(m) where m is the number of elements in a linked list2
3. memory to hold set of list1 = O(len(set1))
4. memory to hold set of list2 = O(len(set2))
3. memory to hold common elements of set1 and set2 = O(k) where k is the number of common elements and in worst case = min(len(set1),len(set2))

hence total space complexity = O(n)+O(m)+O(len(set1))+O(len(set2))+O(k)
